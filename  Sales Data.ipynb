# Handling Missing Values

```python
# Check for missing values

print(df.isnull().sum())
```

# Removing Duplicates

```python
# Check for duplicate rows

print(df.duplicated().sum())
```

# Standardizing Text Values

```python
# Convert column names to lowercase and replace spaces with underscores

df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df.head()
```

```python
# Standardize text columns

df['place_name'] = df['place_name'].str.strip().str.lower()
df['county'] = df['county'].str.strip().str.lower()
df['city'] = df['city'].str.strip().str.lower()
df['state'] = df['state'].str.strip().str.lower()
df['region'] = df['region'].str.strip().str.lower()
```

```python
df.head()
```

# Converting Date Formats

```python
# Convert 'order_date' to datetime format

df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
df['order_date'] = df['order_date'].dt.strftime('%d-%m-%Y')
```

```python
df.head()
```

# Checking and Fixing Data Types

```python
# Ensure 'quantity' is integer

df['qty_ordered'] = df['qty_ordered'].astype(int)
```
